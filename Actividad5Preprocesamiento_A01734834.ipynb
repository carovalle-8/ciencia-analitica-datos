*Asimetría Negativa (o sesgada a la izquierda): Cuando el valor de asimetría es menor que 0, lo que indica que la cola de la distribución se extiende más a la izquierda (valores más bajos).

```python
from scipy.stats import ttest_ind
def Diagnose_MV_Numerical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}
    
    labels=[]
    box_sr = pd.Series('',index = BM_MV.unique())    
    for poss in BM_MV.unique():
        BM = BM_MV == poss
        box_sr[poss] = df[BM][str_att_name].dropna()
        labels.append(MV_labels[poss])
   
    plt.boxplot(box_sr,vert=False)
    plt.yticks([1,2],labels)
    plt.xlabel(str_att_name)
    plt.show()
    
    plt.figure(figsize=(10,4))
    
    att_range = (df[str_att_name].min(),df[str_att_name].max())

    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].hist()
        plt.xlim = att_range
        plt.xlabel(str_att_name)
        plt.title(MV_labels[poss])

    plt.show()
    
    group_1_data = df[BM_MV][str_att_name].dropna()
    group_2_data = df[~BM_MV][str_att_name].dropna()
    
    p_value = ttest_ind(group_1_data,group_2_data).pvalue
    
    print('p-value of t-test: {}'.format(p_value))
```

```python
from scipy.stats import chi2_contingency

def Diagnose_MV_Categorical(df,str_att_name,BM_MV):
    MV_labels = {True:'With Missing Values',False:'Without Missing Values'}

    plt.figure(figsize=(10,4))
    for i,poss in enumerate(BM_MV.unique()):
        plt.subplot(1,2,i+1)
        BM = BM_MV == poss
        df[BM][str_att_name].value_counts().plot.bar()
        plt.title(MV_labels[poss])
    plt.show()

    contigency_table = pd.crosstab(BM_MV,df[str_att_name])   
    p_value = chi2_contingency(contigency_table)[1]
    
    print('p-value of Chi_squared test: {}'.format(p_value))
```

```python
categorical_attributes = ['Gender', 'Education']

BM_MV = response_df.Weight.isna()
for att in categorical_attributes:
    print('Diagnosis Analysis of Missing Values for {}:'.format(att))
    Diagnose_MV_Categorical(response_df,att,BM_MV)
    print('- - - - - - - - - - - - divider - - - - - - - - - - - ')
```

6. Utiliza una medida de tendencia central para imputar los valores faltantes de `Weight`. ¿Cuál es recomendable y por qué?

Mediana:

La mediana es el valor central cuando los datos están ordenados y se utiliza para imputar los valores faltantes.

Ventajas: No se ve afectada por valores atípicos, lo que la convierte en una opción más robusta que la media, especialmente si la distribución de Weight es asimétrica.

Desventajas: Puede no reflejar la tendencia general en distribuciones muy sesgadas.

9. ¿Hay encuestados que sean atípicos según su índice de masa corporal (`BMI`)? El `BMI` puede ser calculado usando la siguiente fórmula:

`BMI = Weight / Height ^ 2`

**NOTA**. El peso tiene que estar en kilogramos y la altura en metros, pero en el conjunto la altura se ha registrado en centímetros, por lo que deberás convertirla a metros.

```python
response_df['Height_m'] = response_df['Height'] / 100  #Altura de cm a m

#Calcular el BMI
response_df['BMI'] = response_df['Weight'] / (response_df['Height_m'] ** 2)

#Calcular IQR para BMI
Q1_BMI = response_df['BMI'].quantile(0.25)
Q3_BMI = response_df['BMI'].quantile(0.75)
IQR_BMI = Q3_BMI - Q1_BMI

#Definir límites 
lower_BMI = Q1_BMI - 1.5 * IQR_BMI
upper_BMI = Q3_BMI + 1.5 * IQR_BMI

#Identificar valores atípicos y contarlos
bmi_atipicos = response_df[(response_df['BMI'] < lower_BMI) | (response_df['BMI'] > upper_BMI)]
num_bmi_atipicos = bmi_atipicos.shape[0]
print(f'Número de valores atípicos en BMI: {num_bmi_atipicos}')

bmi_atipicos_df = bmi_atipicos.copy()

#Eliminar valores atípicos del DataFrame original
response_df_limpio_bmi = response_df[~response_df.index.isin(bmi_atipicos.index)]

# Verificar que los valores atípicos hayan sido eliminados
print(f'Número de registros en el conjunto original: {response_df.shape[0]}')
print(f'Número de registros después de eliminar valores atípicos: {response_df_limpio_bmi.shape[0]}')
```
